```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# keyword = "원피스"
# url = f"https://search.shopping.naver.com/search/all?query={keyword}&frm=NVSHATC&prevQuery={keyword}"
url = 'https://search.shopping.naver.com/search/all?color=16&frm=NVSHPRC&origQuery=원피스&pagingIndex=1&pagingSize=40&productSet=total&query=원피스&sort=rel&timestamp=&viewType=list'

result = requests.get(url, verify=False)
result_text = result.text

soup = BeautifulSoup(result_text, 'html.parser')
```

```python
soup
```

```python
elements = soup.select(".list_basis > div > div .basicList_info_area__TWvzp")
len(elements) # 5

items = []
for element in elements:
    data = {
        "product" : element.select_one(".basicList_title__VfX3c > a").text
#         "price" : element.select_one(".basicList_price_area__K7DDT .price_num__2WUXn").text,
    }
    items.append(data)
```
